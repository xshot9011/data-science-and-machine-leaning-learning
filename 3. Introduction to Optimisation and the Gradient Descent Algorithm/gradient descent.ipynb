## gradient descend as a python function

```python
def gradient_descent(derivative_function, initial_guess, multiplier, precision, n=100):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_function(new_x)]


    for i in range(n):
        previous_x = new_x
        gradient = derivative_function(previous_x)  
        new_x = previous_x - multiplier * gradient 
        
        x_list.append(new_x)
        slope_list.append(derivative_function(new_x))
        
        
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    
    return new_x, x_list, slope_list
```

```python
# plot and derivative
def plot(basic_function, derivative_function, x, x_list, slope_list, 
         x_lim_list, y_lim_list):
    plt.figure(figsize=[15, 6])

    plt.subplot(1, 2, 1)  # row, columns and index
    plt.plot(x, basic_function(x))
    plt.xlim(x_lim_list)
    plt.xlabel('x', fontsize=16)
    plt.ylabel(basic_function.__name__, fontsize=16)
    x_list = np.array(x_list)
    plt.scatter(x_list, g(x_list), color='red', alpha=0.6)

    plt.subplot(1, 2, 2)  # row, columns and index
    plt.grid()  # apply the grid to the chart
    plt.plot(x, derivative_function(x))
    plt.xlim(x_lim_list)
    plt.xlabel('x', fontsize=16)
    plt.ylabel(derivative_function.__name__, fontsize=16)
    plt.scatter(x_list, slope_list, color='red', alpha=0.6)

    plt.show()
```

# diff equal to

$$\frac{\partial f}{\partial x} = \frac{2x \ln(3) \cdot 3^{-x^2 - y^2}}{\left( 3^{-x^2 - y^2} + 1 \right) ^2}$$

$$\frac{\partial f}{\partial y} = \frac{2y \ln(3) \cdot 3^{-x^2 - y^2}}{\left( 3^{-x^2 - y^2} + 1 \right) ^2}$$

```python
from math import log
def partial_dif_x(x, y):
    r = 3**(-x**2 - y**2)
    return 2*x*log(3)*r / (r+1)**2
def partial_dif_y(x, y):
    r = 3**(-x**2 - y**2)
    return 2*y*log(3)*r / (r+1)**2
```

```python
partial_dif_x(1.8, 1.0)
```

```python
multiplier = 0.1
n = 300
param = np.array([1.8, 1])  # initial guess

for i in range(n):
    gradients = np.array([partial_dif_x(param[0], param[1]), 
                          partial_dif_y(param[0], param[1])])
    # learning phase
    param = param - multiplier * gradients  # similar to te old one
    
print('value of param:', param)
print('value of gradients', gradients)
print('cost:', f(param[0], param[1]))

# costom function is much faster than SymPy
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

## pratail dif with respecto $\theta_0$ and $\theta_1$

$$ \frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n}\sum_{i=1}^{n} \big(y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

$$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

```python
def gradient(x, y, theta):
    n = y.size
    
    theta0_slope = (-2/n) * sum(y-theta[0]-theta[1]*x)
    theta1_slope = (-2/n) * sum((y-theta[0]-theta[1]*x)*x)
    
    return np.array([theta0_slope[0], theta1_slope[0]])
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])
# collect data point for scatter plot
plot_values = thetas.reshape(1, 2)
mse_values = get_mse(y_5, thetas[0] + thetas[1] * x_5)

for i in range(1000):
    thetas = thetas - multiplier * gradient(x_5, y_5, thetas)
    
    # append to mse_value
    plot_values = np.concatenate((plot_values, thetas.reshape(1, 2)), axis=0)
    mse_values =  np.append(arr=mse_values, values=get_mse(y_5, thetas[0] + thetas[1] *  x_5))
    
print('min occurs at Theta0 :', thetas[0])
print('min occurs at Theta1 :', thetas[1])
print('MSE is :', get_mse(y_5, thetas[0] + thetas[1] * x_5))
```
