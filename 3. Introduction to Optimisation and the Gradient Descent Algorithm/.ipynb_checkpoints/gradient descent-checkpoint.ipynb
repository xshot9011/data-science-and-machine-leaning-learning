# Notebook imports and Packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example 1 - A sample cose function
$f(x) = x^2 + x + 1$

```python
def f(x):
    return x**2 + x + 1
```

```python
# linspace return evenly spaced numbers over a specificed interval
x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
# plot 
plt.plot(x_1, f(x_1))
plt.xlim(-3, 3)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.show()
```

# derivative of f(x) and slope

```python
def dif(x):
    return 2*x + 1
# at slope = 0 >> 2x + 1 = 0
# x = -1/2
# call f(-1/2) to find value
f(-1/2)
```

```python
# or
min(f(x_1))
# because there's a gap between number
```

```python
# plot and derivative
plt.figure(figsize=[15, 6])

plt.subplot(1, 2, 1)  # row, columns and index
plt.plot(x_1, f(x_1))
plt.xlim(-3, 3)
plt.xlabel('x', fontsize=16)
plt.ylabel('f(x)', fontsize=16)

plt.subplot(1, 2, 2)  # row, columns and index
plt.grid()  # apply the grid to the chart
plt.plot(x_1, dif(x_1))
plt.xlim(-3, 3)
plt.xlabel('x', fontsize=16)
plt.ylabel('dif(x), slope', fontsize=16)

plt.show()
```

# Python loops & gradient descent
gradient descent is going to take place on a cost function

```python
new_x = 3
previous_x = 0
step_multipier = 0.1
n = 30

for i in range(n):
    previous_x = new_x  # make a guess
    gradient = dif(previous_x)  # calculate the error, the more slope is, the more error we got
    new_x = previous_x - step_multipier * gradient  # learning, move against the gradient toward the minimun

# this is machine learning    

print('n:', n)
print('Local minimum occurs at:', new_x)
print('Slope of dif(x) value at this point:', dif(new_x))
print('cost at this point:', f(new_x))
```

```python
new_x = 3
previous_x = 0
step_multipier = 0.1
n = 1000

for i in range(n):
    previous_x = new_x  # make a guess
    gradient = dif(previous_x)  # calculate the error, the more slope is, the more error we got
    new_x = previous_x - step_multipier * gradient  # learning, move against the gradient toward the minimun

# this is machine learning    

print('n:', n)
print('Local minimum occurs at:', new_x)
print('Slope of dif(x) value at this point:', dif(new_x))
print('cost at this point:', f(new_x))
```

```python
new_x = 3
previous_x = 0
step_multipier = 0.1
precision = 0.001

for i in range(100):
    previous_x = new_x  # make a guess
    gradient = dif(previous_x)  # calculate the error, the more slope is, the more error we got
    new_x = previous_x - step_multipier * gradient  # learning, move against the gradient toward the minimun
    
    # limit the precision
    step_size = abs(new_x - previous_x)
    print(step_size)
    if step_size < precision:
        break
        
# this is machine learning    

print('Local minimum occurs at:', new_x)
print('Slope of dif(x) value at this point:', dif(new_x))
print('cost at this point:', f(new_x))
```

```python
new_x = 3
previous_x = 0
step_multipier = 0.1
n = 30

x_list = [new_x]
slope_list = [dif(new_x)]


for i in range(n):
    previous_x = new_x  # make a guess
    gradient = dif(previous_x)  # calculate the error, the more slope is, the more error we got
    new_x = previous_x - step_multipier * gradient  # learning, move against the gradient toward the minimun
    
    x_list.append(new_x)
    slope_list.append(dif(new_x))
    
# this is machine learning    

print('n:', n)
print('Local minimum occurs at:', new_x)
print('Slope of dif(x) value at this point:', dif(new_x))
print('cost at this point:', f(new_x))
```
