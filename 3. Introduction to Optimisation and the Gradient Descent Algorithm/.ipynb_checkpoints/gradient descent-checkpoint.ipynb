# Python loops & gradient descent
gradient descent is going to take place on a cost function

```python
new_x = 3
previous_x = 0
step_multipier = 0.1
n = 30

for i in range(n):
    previous_x = new_x  # make a guess
    gradient = dif(previous_x)  # calculate the error, the more slope is, the more error we got
    new_x = previous_x - step_multipier * gradient  # learning, move against the gradient toward the minimun

# this is machine learning    

print('n:', n)
print('Local minimum occurs at:', new_x)
print('Slope of dif(x) value at this point:', dif(new_x))
print('cost at this point:', f(new_x))
```

# Example 2 - Multiple Minima vs initial Guess & advanced Functions
$g(x) = x^4 - 4x^2 + 5$

```python
x_2 = np.linspace(-2, 2, 1000)

def g(x):
    return x**4 - 4*x**2 + 5
def difg(x):
    return 4*x**3 - 8*x 
```

```python
# plot and derivative
plt.figure(figsize=[15, 6])

plt.subplot(1, 2, 1)  # row, columns and index
plt.plot(x_2, g(x_2))
plt.xlim(-2, 2)
plt.xlabel('x', fontsize=16)
plt.ylabel('g(x)', fontsize=16)

plt.subplot(1, 2, 2)  # row, columns and index
plt.grid()  # apply the grid to the chart
plt.plot(x_2, difg(x_2))
plt.xlim(-2, 2)
plt.xlabel('x', fontsize=16)
plt.ylabel('difg(x), slope', fontsize=16)

plt.show()
```

## gradient descend as a python function

```python
def gradient_descent(derivative_function, initial_guess, multiplier, precision, n=100):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_function(new_x)]


    for i in range(n):
        previous_x = new_x
        gradient = derivative_function(previous_x)  
        new_x = previous_x - multiplier * gradient 
        
        x_list.append(new_x)
        slope_list.append(derivative_function(new_x))
        
        
        step_size = abs(new_x - previous_x)
        if step_size < precision:
            break
    
    return new_x, x_list, slope_list
```

```python
# plot and derivative
def plot(basic_function, derivative_function, x, x_list, slope_list, 
         x_lim_list, y_lim_list):
    plt.figure(figsize=[15, 6])

    plt.subplot(1, 2, 1)  # row, columns and index
    plt.plot(x, basic_function(x))
    plt.xlim(x_lim_list)
    plt.xlabel('x', fontsize=16)
    plt.ylabel(basic_function.__name__, fontsize=16)
    x_list = np.array(x_list)
    plt.scatter(x_list, g(x_list), color='red', alpha=0.6)

    plt.subplot(1, 2, 2)  # row, columns and index
    plt.grid()  # apply the grid to the chart
    plt.plot(x, derivative_function(x))
    plt.xlim(x_lim_list)
    plt.xlabel('x', fontsize=16)
    plt.ylabel(derivative_function.__name__, fontsize=16)
    plt.scatter(x_list, slope_list, color='red', alpha=0.6)

    plt.show()
```

```python
local_min, x_list, slope_list = gradient_descent(difg, 0.5, 0.02, 0.001)
print('local minimun:', local_min)
print('len(x_list):', len(slope_list))

plot(g, difg, x_2, x_list, slope_list, [min(x_2), max(x_2)], [min(g(x_2)), max(g(x_2))])
```

```python
local_min, x_list, slope_list = gradient_descent(difg, -0.5, 0.02, 0.001)
print('local minimun:', local_min)
print('len(x_list):', len(slope_list))

plot(g, difg, x_2, x_list, slope_list, [min(x_2), max(x_2)], [min(g(x_2)), max(g(x_2))])
```

# Example 3 - Divergence, Overflow and Python Tuples
$h(x) = x^5 - 2x^4 + 2$

```python
x_3 = np.linspace(-2.5, 2.5, 1000)

def h(x):
    return x**5 - 2*x**4 + 2

def difh(x):
    return 5*x**4 - 8*x**3
```

```python
local_min, x_list, slope_list = gradient_descent(difh, 0.5, 0.02, 0.0001, n=1000)

plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)

plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)
plt.scatter(x_list, h(np.array(x_list)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(x_3, difh(x_3), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

plt.show()

print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
print('Number of steps: ', len(x_list))
```

```python
# if you re run with initial = -0.2 
# you will got an overlow error
# our algorithm with go down and down on left hand side
# let try witht initial=-0.2 and n=1000
# the decrease the n value 65 >> 66 >> 67 >> 68 >> 71 >> 73 (overflow problem)
# notice the local_min and the cost
```

```python
local_min, x_list, slope_list = gradient_descent(difh, -0.2, 0.02, 0.0001, n=73)

plt.figure(figsize=[15, 5])

# 1 Chart: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)

plt.title('Cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)

plt.plot(x_3, h(x_3), color='blue', linewidth=3, alpha=0.8)
plt.scatter(x_list, h(np.array(x_list)), color='red', s=100, alpha=0.6)

# 2 Chart: Derivative
plt.subplot(1, 2, 2)

plt.title('Slope of the cost function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(x_3, difh(x_3), color='skyblue', linewidth=5, alpha=0.6)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

plt.show()

print('Local min occurs at: ', local_min)
print('Cost at this minimum is: ', h(local_min))
print('Number of steps: ', len(x_list))
```
